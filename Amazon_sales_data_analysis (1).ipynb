```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# %matplotlib inline is a command that enables Matplotlib plots to be displayed inline in Jupyter Notebook, instead of opening a separate window.
%matplotlib inline
```

```python
# Saving the data dataframe to df
df = data
```

#Now let's create heatmap to find correlation between features.

```python
sns.heatmap(df.corr(), annot=True)
```

#Observations:
* As you can see in the above heatmap Unit Price and Unit Cost are Highly corelated.
* Also Unit Price moderately related to Total Revenue and Total Cost but less related to Total Profit.
*  No relation between Order ID and rest other features.

#Region

* Total Revenue by region

```python
df.groupby('Region')['Total Revenue','Total Profit'].sum().plot(kind = 'bar', color = ['Yellow','Blue'])
plt.title('Total Revenue By Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue')
plt.show()
```

* Sub-saharan Africa has the highest revenue and profit.

```python
sns.barplot(x= 'Order Priority', y= 'Total Profit',color = 'orange', data =df)
plt.show()
```

```python
item_type = df['Item Type'].value_counts().index
item_val = df['Item Type'].value_counts().values
# Pie Chart for item
fig,ax = plt.subplots(figsize=(9,9))
ax.pie(x=item_val[:],labels=item_type[:],autopct='%1.2f%%')
plt.title('Which Item Covers Most Area')
plt.show()
```

```python
df['Order Priority'].value_counts()
```

```python
Order_Priority = df['Order Priority'].value_counts().index
Order_Priority_val = df['Order Priority'].value_counts().values
# Pie Chart for item
fig,ax = plt.subplots(figsize=(9,9))
ax.pie(x=Order_Priority_val[:],labels=Order_Priority[:],autopct='%1.2f%%')
plt.title('Order_Priority')
plt.show()
```

```python
df.groupby('Item Type').mean()['Units Sold'].plot(kind = 'bar',color = 'Green')
plt.title('Avg. Units sold by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Avg. Units Sold')
plt.show()
```

```python
# Calculating the total profit for each group with respect to Item Type and then sorting then in descending order
profit_by_category = df.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)
profit_by_category.plot(kind='bar')
```

#Analysing Unit Price and Revenue

```python
plt.scatter(df['Unit Price'],df['Total Revenue'],color = 'Blue')
plt.title('Unit Price Vs Total Revenue')
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
```

# As obvious, revenue increases as unit price increses.

** converting dates to datetime format **

#Total Revenue overtime

```python
df['Order Date']= pd.to_datetime(df['Order Date'])
```

```python
df.groupby('Order Date').sum()['Total Revenue'].plot(kind = 'line',color = 'Blue')
plt.title('Total Revenue over time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
print('*'*20,'Total Revenue:',df['Total Revenue'].sum(),'*'*23)
plt.show()
```

#Monthly sale

```python
df['Order Month']=df['Order Date'].dt.month
```

```python
df.groupby('Order Month').sum()[['Total Revenue']].plot(kind='bar')
plt.title('Total Revenue By Month')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

#Yearly sale

```python
df['Order Year']=df['Order Date'].dt.year
```

```python
df.groupby('Order Year').sum()['Total Revenue'].plot(kind = 'bar', color = 'Green')
plt.title('Total Revenue By Year')
plt.xlabel('Order Year')
plt.ylabel('Total Revenue')
plt.show()
```

# Yearly Monthwise sales

```python
Yearly_Monthwise = df.groupby(['Order Month','Order Year']).sum().reset_index()
Yearly_Monthwise.head()
```

```python
sns.boxplot(x ='Order Month', y = 'Total Profit', data = Yearly_Monthwise)
```
